```python
import numpy as np
import matplotlib.pyplot as plt
from tabula import read_pdf
df = read_pdf('https://www3.akindo-sushiro.co.jp/pdf/menu/allergy.pdf',pages='all',multiple_tables=True,lattice=True)
df
```

```python
df[0].columns
```

```python
takhalal = []
for i in range(np.size(df)):
    df_page = df[i]
    beef = df_page['メニュー名称']['●' == df_page['牛\r肉']]
    chicken = df_page['メニュー名称']['●' == df_page['鶏\r肉']]
    pork = df_page['メニュー名称']['●' == df_page['豚\r肉']]
    gelatine = df_page['メニュー名称']['●' == df_page['ゼ\rラ\rチ\rン']]
    combined_array = np.concatenate((beef,chicken,pork,gelatine))
    unique_variables = np.unique(combined_array)
    for j in range(np.size(unique_variables)):
        takhalal.append(unique_variables[j])
```

```python
for i in range(np.size(takhalal)):
    print(takhalal[i])
```
